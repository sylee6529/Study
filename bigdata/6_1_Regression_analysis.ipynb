# 빅데이터 회귀 분석

Sikit Learn 사용법

```python
import matplotlib.pyplot as plt
import numpy as np

rng = np.random.RandomState(42)
x = 10 * rng.rand(50)
y = 2 * x - 1 + rng.randn(50)
plt.scatter(x, y)
```

```python
# 모델 클래스 선택
from sklearn.linear_model import LinearRegression
# 모델 초모수 선택
model = LinearRegression(fit_intercept=True) #fit_intercept 는 절편 사용 여부
```

```python
# Sciket-Learn 에서는 특징 행렬은 2차원으로, 대상 행렬은 1차원으로 표현되어야 한다

# 1차원으로 생성된 데이터를 2차원으로 변경하는 법1
X = x[:, np.newaxis]
X.shape

# 1차원으로 생성된 데이터를 2차원으로 변경하는 법2
X = x.reshape(50, 1)
X.shape
```

```python
# 모델에 데이터 적용
model.fit(X, y) # 계산 결과는 모델 인스턴스 속성에 저장됨

# 모델 확인
print(model.coef_)
print(model.intercept_)

# 모델 정확도 확인 (1에 가까울 수록 정확하다)
model.score(X, y)
```

```python
xfit = np.linspace(-1, 11)
plt.scatter(x, y)
plt.plot(xfit, model.coef_*xfit + model.intercept_, '-r')
```

현실적으로 데이터가 선형으로 표현되지 않는 경우에는 간단한 선형 회귀 모델의 정확도가 많이 떨어진다.
