1) 선그림(Line plot)

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
```

```python
fig = plt.figure() # 축, 그래픽, 텍스트, 레이블을 표시하는 모든 객체를 포함하는 컨테이너
fig.suptitle("Example1:LinePlot")
ax = plt.axes() # 눈금과 레이블이 있는 테두리 상자(데이터가 표시되는 부분)
```

```python
x = np.linspace(0,10,25)
plt.plot(x,x,'o', color='r')
plt.plot(x,x+2,'.',color='g')
plt.plot(x,x+4,'s-',color='b')
plt.plot(x,x+6,'d--',color='c')
```

```python
plt.plot(x, np.cos(x * 2))
plt.axis([-1, 11, -1.5, 1.5])
```

```python
plt.plot(x, np.exp(-x));
plt.yscale("log")
```
