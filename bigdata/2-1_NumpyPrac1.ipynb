### - 데이터 분석

- 데이터는 다양한 형식으로 존재함
    - Ex. 이미지 데이터 → 2차원 숫자배열(픽셀 밝기)
    - 사운드 데이터 → 1차원 배열(시간대비 강도)
    - 
- (과정)원본 데이터를 숫자 배열로 변환 → 편한 방식으로 연산 적용 → 연산 결과를 다루기 쉬운 형태로 변환
- 숫자 배열을 효율적으로 저장하고 가공하는 것은 빅데이터 분석을 위한 기본적인 절차이다. (Python-Numpy 패키지 이용)

### - 넘파이 (Numerical Python)

- array 단위로 데이터 관리 및 연산 수행

1) 넘파이 패키지 임포트

```python
import numpy as np
```

```python
data1 = [1, 2, 3, 4, 5]
data2 = [6, 5, 4, 3, 2, 1]

print(data1, data2)
```

```python
ar1 = np.array(data1)
ar1
```

```python
ar2 = np.array(data1, dtype='float32')
ar2
```

```python
ar3 = np.array([range(i, i + 3) for i in data1])
ar3
```

```python
ar4 = np.full((3, 5), 3.14)
ar4
```

```python
ar5 = np.arange(0, 20, 2)
ar5
```

```python
ar6 = np.linspace(0, 1, 3)
ar6
```

```python
ar7 = np.random.random((3, 3))
ar7
```

```python
ar8 = np.random.randint(0, 100, (3, 3))
ar8
```

```python
ar9 = np.eye(3)
ar9
```

```python
x = np.random.randint(10, size=(3, 4, 5))
x
```

```python
print("x ndim: ", x.ndim)
print("x shape: ", x.shape)
print("x size: ", x.size)
```
